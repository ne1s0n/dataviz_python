```python
my_years = [1980, 1990, 2000, 2010]
df_years = df_canada.loc[:, my_years].transpose().reset_index()
df_years.rename(columns = {'index':'Years'}, inplace = True)

#from wide to long
df_years_long = df_years.melt(id_vars = 'Years', var_name='Country', 
                              value_name='Immigrants')


#ready for a barplot
sns.barplot(x = 'Years', y = 'Immigrants', data = df_years_long)
```
