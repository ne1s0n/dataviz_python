Take a look to the [list methods available in python](https://www.w3schools.com/python/python_lists_methods.asp) and then

* Append the second list to the first one, so that now it contains eight values
* extract the count of how many times in the list appears the value `3.14`

---

```python
# Your solution here

first = [1, 2, 4, 8]
second = [3.14, 88, 3.14, 101]
print(first)
print(second)

#be sure to not mix .append() and .extend()
first.extend(second)
print(first)

print(first.count(3.14))

first.sort(reverse = True)
print(first)
```

## ASSIGNMENT! (dictionary)

Take a look to the [dictionary methods available in python](https://www.w3schools.com/python/python_dictionaries_methods.asp) and then

* declare a dictionary with four keys, one for each season. The values may be whatever you prefer
* print the list of keys and the list of values
* remove the entry for summer

---

```python
# Your solution here
myseasons = {
    'spring' : 'sneezing',
    'summer' : 'sweating',
    'autumn' : 'THE MOST BEAUTIFUL',
    'winter' : 'is coming'
}

print(myseasons.keys())
print(myseasons.values())

print(myseasons)
myseasons.pop('summer')
print(myseasons)
```

The remainder operator ``%`` returns the remainder of a division, e.g. ``13 % 5`` would return 3.

Use it to write some code that prints all the prime numbers under 100. 

---

---

```python

# A very simple, very inefficient implementation

#looping over all integers between two and one hundred
for i in range(2, 100):
  #setting a flag
  is_prime = True
  
  #we need to check our current teste value i against
  #all smaller integers 
  for j in range(2, i-1):

    #checking if the remainder of the division is zero
    if i % j == 0:
      is_prime = False
  
  #here we are still inside the outer loop. Printing in case
  #of a prime number
  if is_prime:
    print (i)
```
