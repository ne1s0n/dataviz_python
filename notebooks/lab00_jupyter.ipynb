Jupiter notebooks are increasingly popular tools used to mix text and code. They are ideal for compiling report and ETA (Exploratory Data Analysis). 

Jupiter notebooks are made by sequential "blocks", called cells. There are two main type of cell: text and code. This very text you are reading is inside a text cell...

## What is a text cell

A text cell contain the instruction to render graphically a piece of text. To do so it used a formatting language called `markdown` that lets you easily implement most of the standard operation. You can have **bold text**, *italic text*, `monospace font`, $math \ symbols$ and so forth.

### You can also have subtitles

#### And subsubtitles

##### And even subsubsubtitles

Other stuff easily implemented in `markdown`:

* unordered and ordered lists
* links
* horizontal lines

When in doubt it's easy to search for compact guides (often called cheatsheet) that contain all the conventions in a handy place. Try for example [this one](https://blog.sqlbak.com/wp-content/uploads/2020/12/Jupyter-Notebook-Markdown-Cheatsheet2.pdf) for a compact option.

Also, you can always copy what other people is doing. Try and double click on this very cell and see how I obtained this formatting.

## Embedding HTML

The whole point of using `markdown` is to NOT have to write HTML directly. That said, some formatting is simply absent and it cannot be achieved without direct html. One simple example is having a centered text, for which you must use the ``<p>`` tag with the proper ``align`` property to obtain

<p align="center">this awesome result</p>

Keep in mind that HTML is a mess and there's a bunch of stuff that is still supported by browsers but deprecated by HTML5. 

To explore the state of the art on the topic just reference [w3school website](https://www.w3schools.com/tags/default.asp).

## What about CSS

Jupyter notebooks support CSS, which is also the cleanest way to format HTML. But! We are working in google colab, which have a complicated relationship with css. As an example to change the color of your font you should use:

```
<p style="color: red">This is a red paragraph.</p>

```
But if you try you'll see that it doesn't work:

<p style="color: red">This is a red paragraph.</p>

Take this as a cautionary tale to keep your formatting simple.
